````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
%matplotlib inline
````

````python
# Váš kód zde


########################################################
# Předpřipravená kostra modelu náhodného lesa
class CustomRandomForest:
    """
    Třída Vašeho modelu
    Bude se jednat o model náhodného lesa, kde podmodely tvoří rozhodovací stromy pro regresi.
    Pro podmodely můžete použít implementaci DecisionTreeRegressor ze sklearn.
    """
    def __init__(self, n_estimators, max_samples, max_depth, **kwargs):
        """
        Konstruktor modelu
        Základní hyperparametery:
            n_estimators - počet podmodelů - rozhodovacích stromů.
            max_samples - vyberte si, zda tento parametr bude označovat relativní počet bodů (tj. číslo mezi 0 a 1) 
                          nebo absolutní počet bodů (tj. číslo mezi 1 a velikostí trénovací množiny), 
                          které budou pro každý podmodel rozhodovacího stromu náhodně vybrány z trénovací množiny (bootstrap) a použity k jeho trénování.
            max_depth - maximální hloubka každého z podmodelů rozhodovacího stromu.
            kwargs - (volitelně) případné další hyperparametry, které pošlete do podmodelů rozhodovacího stromu
        """
        self.n_estimators = n_estimators
        self.max_samples = max_samples
        self.max_depth = max_depth
    def fit(self, X, y):
        """
        Natrénování modelu. Trénovací data jsou v argumentech X a y.
        Pro trénování podmodelů používejte bootstraping a velikost samplovaného vzorku vezměte z hyperparametru max_samples_fraction
        """
        self.forest = []
        self.forest_feachure = []
        for i in range(self.n_estimators):
            row_lines = np.random.randint(0,X.shape[0],size = int(X.shape[0] * self.max_samples))
            column_lines = np.random.choice(range(0,X.shape[1]),size = X.shape[1] // 3 ,replace=False)
            x_sample = X.iloc[row_lines,column_lines]
            y_sample = y.iloc[row_lines]
            reg = DecisionTreeRegressor(max_depth = self.max_depth,criterion="squared_error")
            reg.fit(x_sample,y_sample)
            self.forest.append(reg)
            self.forest_feachure.append(column_lines)
    def predict(self, X):
        predictions = [tree.predict(X.iloc[:,self.forest_feachure[index]]) for index,tree in enumerate(self.forest)]
        final_prediction = np.mean(predictions, axis=0)
        return final_prediction
````

````python
#Readings and splitting the dataset into parts
random_seed = 42
df = pd.read_csv("data.csv")
Xtrain,Xrest,ytrain,yrest = train_test_split(df.drop(columns = ["Life expectancy"]),df["Life expectancy"],test_size=0.4,random_state=random_seed)
Xval,Xtest,yval,ytest = train_test_split(Xrest,yrest,test_size=0.4,random_state=random_seed)
evaluation = pd.read_csv("evaluation.csv")
````

````python
display(Xtrain.select_dtypes("object"))
display(Xtrain["Status"].unique())
````

### Explanation of the Code

1. **Copying Datasets**  
   The datasets `Xtrain`, `Xval`, and `Xtest` are copied to prevent unintended modifications to the original data.

2. **Saving Country Names**  
   The `Country` column from the `evaluation` dataset is stored for later use after transformations.

3. **Batch Dataset Processing**  
   All datasets (`Xtrain`, `Xval`, `Xtest`, `evaluation`) are grouped into a list to apply transformations uniformly.

4. **Defining 'Status' as Categorical**  
   The `Status` column is converted into an ordered categorical type ("Developing" < "Developed") to preserve its hierarchy.

5. **Encoding Categorical Variables**  
   The `Status` and `Country` columns are encoded into numeric codes to make the data compatible with machine learning models.

6. **Updating Transformed Datasets**  
   The transformed datasets are unpacked back into `Xtrain`, `Xval`, `Xtest`, and `evaluation`.

---

### Purpose

- **Model Compatibility**: Converts categorical data into numeric format suitable for training.  
- **Consistency**: Ensures all datasets are encoded in the same manner.  
- **Preservation**: Keeps the original `Country` names for potential use in interpretation.  

````python
Xtrain = Xtrain.copy()
Xval = Xval.copy()
Xtest = Xtest.copy()
country_name = evaluation["Country"]
datasets = [Xtrain, Xval, Xtest,evaluation]
status_category = pd.CategoricalDtype(categories=["Developing", "Developed"], ordered=True)
for i in range(len(datasets)):
    datasets[i]["Status"] = datasets[i]["Status"].astype(status_category)
    datasets[i]["Country"] = datasets[i]["Country"].astype("category")
    datasets[i]["Status"] = datasets[i]["Status"].cat.codes
    datasets[i]["Country"] = datasets[i]["Country"].cat.codes
Xtrain,Xval,Xtest,evaluation  = datasets
````

### Explanation of the Code

1. **Calculate Median**  
   The median values of all numerical columns (excluding the first three) in the `Xtrain` dataset are computed. This will be used to impute missing values.

2. **Dropping Rows with Critical Missing Data**  
   For each dataset, rows with missing values in the essential columns (`Country`, `Status`, `Year`) are removed to ensure the data remains valid and meaningful.

3. **Filling Missing Values**  
   Remaining missing values in all datasets are replaced with the corresponding median values calculated from the `Xtrain` dataset. This ensures consistency and prevents information leakage.

4. **Update Datasets**  
   The processed datasets are reassigned to their respective variables: `Xtrain`, `Xval`, `Xtest`, and `evaluation`.

---

### Purpose

- **Robustness**: Ensures that key columns do not have missing values, which might disrupt the model.  
- **Uniformity**: Imputes missing values with `Xtrain` medians to avoid test-time information leakage.  
- **Dataset Integrity**: Handles missing data consistently across all datasets to prevent errors in modeling.


````python
train_median = Xtrain.iloc[:,3:].median()
for i in range(len(datasets)):
    datasets[i] = datasets[i].dropna(subset=["Country", "Status", "Year"])
    datasets[i] = datasets[i].fillna(train_median)
Xtrain,Xval,Xtest,evaluation  = datasets
````

### Explanation of the Code

1. **Standard Scaling**  
   - `StandardScaler` is used to normalize features by removing the mean and scaling to unit variance.  
   - The scaler is fit on `Xtrain` and then applied to transform `Xtrain`, `Xval`, `Xtest`, and `evaluation`.  
   - Outputs: `Xtrain_standart`, `Xval_standart`, `Xtest_standart`, and `evaluation_standart`.

2. **Min-Max Scaling**  
   - `MinMaxScaler` scales each feature to a given range (default is [0, 1]).  
   - The scaler is fit on `Xtrain` and then applied to `Xtrain`, `Xval`, and `Xtest`.  
   - Outputs: `Xtrain_scalled`, `Xval_scalled`, and `Xtest_scalled`.

---

### Purpose

- **StandardScaler**: Ensures features have zero mean and unit variance, suitable for models sensitive to feature scale.  
- **MinMaxScaler**: Scales features to a fixed range.  
- **Separate Transformations**: Prevents information leakage by fitting scalers only on the training data.


````python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
scaller = StandardScaler()
Xtrain_standart = scaller.fit_transform(Xtrain)
Xval_standart = scaller.transform(Xval)
Xtest_standart = scaller.transform(Xtest)
evaluation_standart = scaller.transform(evaluation)
scaller = MinMaxScaler()
Xtrain_scalled = scaller.fit_transform(Xtrain)
Xval_scalled = scaller.transform(Xval)
Xtest_scalled = scaller.transform(Xtest)
````

### Explanation of the Code

#### Function: `model_tune`

1. **Parameter Grids**  
   - Defines parameter combinations for tuning based on the model type (`knn`, `forest`, or `ridge`).  
   - Uses `ParameterGrid` to generate all possible combinations of parameters.

2. **Model Selection**  
   - For each parameter combination, a model (`KNeighborsRegressor`, `CustomRandomForest`, or `Ridge`) is initialized and trained using `Xtrain_data` and `ytrain`.

3. **Evaluation**  
   - Predictions are made on `Xval_data`, and evaluation metrics (`root_mean_squared_error` and `mean_absolute_error`) are calculated.  
   - Tracks the best parameter sets for both RMSE and MAE.

4. **Logging Results**  
   - Prints the best RMSE and MAE scores with corresponding parameters.  
   - Records RMSE and MAE for all parameter combinations for visualization.

5. **Purpose**  
   - Finds optimal parameters by testing all combinations to improve model performance on validation data.

---

#### Function: `draw_score`

1. **Plot Creation**  
   - Visualizes RMSE and MAE values for each parameter combination using a line plot.  
   - Blue line represents RMSE, red line represents MAE.

2. **Labels and Legend**  
   - Adds titles, axis labels, and a legend to make the plot informative.

````python
from sklearn.model_selection import ParameterGrid
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import root_mean_squared_error,mean_absolute_error 
from sklearn.linear_model import Ridge
def model_tune(model_type,Xtrain_data,Xval_data):
    if model_type == "knn":
        parametr_grid = {
            "n_neighbors" : range(1,20 + 1),
            'p' : [1,2],
            "weights" : ["uniform","distance"]
        }
    elif model_type == "forest":
        parametr_grid = {
        "n_estimators" : range(1,101,10),
        "max_depth" : range(1,8),
        "max_samples" : np.arange(0.1, 1.1, 0.1) 
        }
    elif model_type == "ridge":
        parametr_grid = {
            "alpha" : np.logspace(-6, 3, num=20),
            "fit_intercept" : [True, False]
        }
    combinations = ParameterGrid(parametr_grid)
    rmse = []
    mae = []
    best_score_rmse = None
    best_score_mae = None
    best_params_rmse = None
    best_params_mae = None
    for i in range(len(combinations)):
        if(model_type == "knn"):
            reg = KNeighborsRegressor(**(combinations[i]))
        elif model_type == "forest":
            reg = CustomRandomForest(**(combinations[i]))
        elif model_type == "ridge":
            reg = Ridge(**(combinations[i]))
        reg.fit(Xtrain_data,ytrain)
        prediction = reg.predict(Xval_data)
        curr_rmse = root_mean_squared_error(yval,prediction)
        curr_mae = mean_absolute_error(yval,prediction)
        if(i == 0 or curr_rmse < best_score_rmse):
            best_score_rmse = curr_rmse
            best_params_rmse = combinations[i]
        if(i == 0 or curr_mae < best_score_mae):
            best_score_mae = curr_mae
            best_params_mae = combinations[i]
        rmse.append(curr_rmse)
        mae.append(curr_mae)
    print(f"Found best RMSE score {best_score_rmse} \nWith  parametrs {best_params_rmse}")
    print(f"Found best MAE score {best_score_mae} \nWith  parametrs {best_params_mae}")

    draw_score(len(combinations),rmse,mae)
    
def draw_score(indexes_len,rmse,mae):
    figure = plt.figure(figsize = (16,8))
    ax = figure.add_subplot()
    ax.plot(range(indexes_len),rmse,color = 'blue',label = "RMSE plot")
    ax.plot(range(indexes_len),mae,color = 'red',label = "MAE plot")
    ax.set_title("Score distribution")
    ax.set_xlabel("Index of combination")
    ax.set_ylabel("Score")
    ax.grid(axis = 'y')
    ax.legend(loc = "best",fontsize = 15)
````

### Model Tuning Strategy

We use the `model_tune` function to tune three models: **K-Nearest Neighbors (KNN)**, **Random Forest (Forest)**, and **Ridge Regression**. The tuning process is performed with different data preprocessing strategies for each model based on their characteristics and performance.

#### 1. **KNN (K-Nearest Neighbors)**
- **Data Used**: Both **standardized** and **scaled** data.
- **Reason**: KNN is sensitive to the scale of the data because it relies on distance metrics (Euclidean, Manhattan, etc.) to determine the neighbors. 

#### 2. **Random Forest (Forest)**
- **Data Used**: **Regular (raw) data**.
- **Reason**: Random Forest is a tree-based algorithm, which is **not sensitive** to feature scaling. Trees split the data based on thresholds, so feature scaling doesn't significantly affect model performance. Therefore, we use the regular data for the Random Forest model to keep it simple and avoid unnecessary computations.

#### 3. **Ridge Regression**
- **Data Used**: **Standardized** and **scaled** data.
- **Reason**: Ridge Regression applies regularization, and the regularization term (`alpha`) is sensitive to the scale of the features. Standardizing or scaling the data ensures that the regularization works correctly, as features with different scales could disproportionately affect the model's coefficients.


````python
model_tune("knn",Xtrain,Xval)
````

````python
model_tune("knn",Xtrain_scalled,Xval_scalled)
````

````python
model_tune("knn",Xtrain_standart,Xval_standart)
````

### KNN Model Results (Standardized, Scaled, and Regular Data)

The KNN model performs significantly better when data is standardized or scaled, as seen in the improved RMSE and MAE scores. For standardized data, the best results are achieved with 3 neighbors and distance-based weighting. The MAE also improves with the use of distance weighting. In general, scaling or standardizing the data optimizes KNN performance, with the optimal number of neighbors being around 2-3 depending on the metric used.


````python
model_tune("forest",Xtrain,Xval)
````

### Random Forest Model Results (Regular Data)

For the Random Forest model using regular (non-scaled) data, the best performance is achieved with 21 estimators, a maximum depth of 7, and using all available samples (max_samples = 1.0). Both the RMSE and MAE scores are optimized with these parameters, indicating that this configuration provides the most accurate predictions on the validation set.


````python
model_tune("ridge",Xtrain_scalled,Xval_scalled)
````

````python
model_tune("ridge",Xtrain_standart,Xval_standart)
````

### Ridge Regression Model Results

For the Ridge Regression model, the best results were found with scaled and standardized data. The RMSE and MAE scores were optimized with specific alpha values (0.002 for RMSE and 0.006 for MAE in scaled data, and 1.438 for both metrics in standardized data). We did not use regular data for Ridge regression, as it provided worse performance and generated a warning due to matrix non-regularity issues.


### Explanation of the Code

#### Step 1: Train the Final Model
- **Model**: A K-Nearest Neighbors (KNN) regressor is initialized with:
  - `weights="distance"`: Assigns closer neighbors higher weight in the prediction.
  - `p=1`: Uses the Manhattan distance (L1 norm).
  - `n_neighbors=3`: The model uses 3 nearest neighbors for prediction.
- **Training**: The model is trained on the standardized training data (`Xtrain_standart`) with the corresponding target values (`ytrain`).

#### Step 2: Evaluate on Validation Data
- **Prediction**: The model makes predictions on the standardized validation data (`Xval_standart`).
- **Metrics**:
  - **RMSE**: Root Mean Squared Error is calculated to evaluate the model's accuracy.
  - **MAE**: Mean Absolute Error is calculated as an additional evaluation metric.
- **Display**: Both RMSE and MAE are displayed for the validation set.

#### Step 3: Evaluate on Test Data
- **Prediction**: The model predicts on the standardized test data (`Xtest_standart`).
- **RMSE**: Test RMSE is calculated to assess the model’s performance on unseen data.
- **Display**: The RMSE result for the test data is displayed.

#### Step 4: Predictions on Evaluation Data
- **Prediction**: The final model makes predictions on the evaluation set (`evaluation_standart`).
- **Round Predictions**: The predicted life expectancy values are rounded to one decimal place.
- **Save Results**: The predictions are saved in a DataFrame along with the corresponding country and year information. The results are then written to a CSV file named `"results"`.

````python
final_reg = KNeighborsRegressor(weights = "distance", p = 1, n_neighbors = 3)
final_reg.fit(Xtrain_standart,ytrain)
valid_prediction  =  final_reg.predict(Xval_standart)
final_rmse = root_mean_squared_error(yval,valid_prediction)
final_mae = mean_absolute_error(yval,valid_prediction)
display(final_rmse)
display(final_mae)
````

````python
test_prediction = final_reg.predict(Xtest_standart)
test_rmse = root_mean_squared_error(ytest,test_prediction)
display(test_rmse)
````

````python
evaluation_prediction = final_reg.predict(evaluation_standart)
evaluation_prediction = np.round(evaluation_prediction, 1)
evaluation_df = pd.DataFrame({"Country" : country_name,"Year" : evaluation["Year"],"Life expectancy" : evaluation_prediction})
evaluation_df.to_csv("results.csv",index = False)
````
