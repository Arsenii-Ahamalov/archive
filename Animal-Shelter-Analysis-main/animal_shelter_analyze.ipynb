```python
#Maybe some missing values are represented in a non-standard way, let's check it and repace them.
def replace_non_standard_missing_data(df):
    # Define patterns representing missing values
    missing_value_patterns = ["?", "-", "None", "", "NA", "Unknown"]
    # Replace these patterns in each column with the specified replacement value
    for col in df:
        df[col] = df[col].replace(missing_value_patterns, np.nan)
```

Now let's translate the types of features to those suitable for our purposes. And we will start with the numeric feature Age upon Intake/Outcome.

```python
intakes_data["Age upon Intake"].unique()
```

It should be noted that there are values with negative time, most likely we are talking about errors and we can replace them with np.nan.Besides, I think it is acceptable to convert all values of years/months/weeks into days to unify the data.Let's simplify that year = 365 days and month = 30 days. 

```python
import re
def transforming_age_column(record):            
    
        if pd.isna(record):
            return np.nan
        
        result = re.split(r'(\d+)', record)
        result = [item.strip() for item in result if item]

        if len(result) != 2 :
            return np.nan
        number = int(result[0])
        unit = result[1].lower()

        if "year" in unit:
            return number * 365
        elif "month" in unit:
            return number * 30
        elif "week" in unit:
            return number * 7
        else:
            return number
```

```python
# Transform Age upon Intake/Outcome to int64
intakes_data["Age upon Intake"] = intakes_data["Age upon Intake"].apply(transforming_age_column)
intakes_data["Age upon Intake"] = intakes_data["Age upon Intake"].apply(lambda x: round(x) if pd.notna(x) else x).astype('Int64')
outcomes_data["Age upon Outcome"] = outcomes_data["Age upon Outcome"].apply(transforming_age_column)
outcomes_data["Age upon Outcome"] = outcomes_data["Age upon Outcome"].apply(lambda x: round(x) if pd.notna(x) else x).astype('Int64')
```

```python
#Transform of all necessary data into datetime format
intakes_data["DateTime"] = pd.to_datetime(intakes_data["DateTime"], format="%m/%d/%Y %I:%M:%S %p")
outcomes_data["DateTime"] = pd.to_datetime(outcomes_data["DateTime"], format="%m/%d/%Y %I:%M:%S %p")
outcomes_data["Date of Birth"] = pd.to_datetime(outcomes_data["Date of Birth"],format = "%m/%d/%Y")
display(intakes_data.info())
display(outcomes_data.info())
```

### Age Upon Intake Analysis

In this analysis, we will explore the age of animals at intake to better understand their distribution. We will utilize two types of visualizations:

1. **Box Plot**: This will provide statistical metrics, including the median, quartiles, and potential outliers.

2. **Hist Plot**: This will illustrate the distribution of age, showing the density of data points across different age groups.

These visualizations will enable us to conduct a thorough statistical analysis of age upon intake.
We will use a log scale to handle the wide range of ages.

```python
age_in_days = intakes_data["Age upon Intake"].dropna() 
age_in_days = age_in_days[age_in_days > 0]  

age_in_days_log = np.log10(age_in_days)

figure = plt.figure(figsize=(16, 8))
ax = figure.add_subplot(1, 2, 1)
ax.boxplot(age_in_days_log, patch_artist=True, 
           boxprops=dict(facecolor='pink', color='red'),
           medianprops=dict(color='white', linewidth=2))
ax.set_title('Box Plot of Age Upon Intake ', fontsize=16)
ax.set_ylabel('Age in Days', fontsize=14)
ax.set_yticks([np.log10(1), np.log10(10), np.log10(100), np.log10(1000), np.log10(10000)])
ax.set_yticklabels([1, 10, 100, 1000, 10000]) 
ax.grid(axis = 'y')

ax2 = figure.add_subplot(1,2,2)
ax2.hist(age_in_days_log,  color='lightblue', edgecolor='black')
ax2.set_title('Histogram of Age Upon Intake')
ax2.set_xlabel('Age in Days')
ax2.set_ylabel('Frequency')
ax2.set_xticks([np.log10(1), np.log10(10), np.log10(100), np.log10(1000), np.log10(10000)])
ax2.set_xticklabels([1, 10, 100, 1000, 10000]) 

ax2.grid(axis = 'y')
```

### Conclusion:

The box plot and histogram show a wide range of animal ages upon intake. ## Exploring the Relationship Between **Intake Type** and **Intake Condition**

We'll analyze the relationship between **Intake Type** (how an animal arrives at the shelter) and **Intake Condition** (the animal's health upon arrival).
This can help identify patterns, like whether strays are more likely to arrive in poor condition compared to owner-surrendered animals.First, let's look at the presence of correlation.

```python
contingency_table = pd.crosstab(intakes_data['Intake Condition'], intakes_data['Intake Type'])
# Convert the contingency table into percentages by dividing each value by the sum of its corresponding row.
# We use percentages instead of absolute counts to compare how intake types distribute across different Intake Condition,
# regardless of the number of occurrences for each intake type.
contingency_percent = contingency_table.div(contingency_table.sum(axis=1), axis=0).round(3)

display(contingency_percent)
plt.figure(figsize=(12, 8))
sns.heatmap(contingency_percent, annot=True, fmt='.2f', cmap='YlGnBu', cbar=True)
plt.title('Heatmap of Intake Conditions and Intake Type')
plt.xlabel('Intake Type')
plt.ylabel('Intake Condition')
plt.show()
```

### Conclusions from the Heatmap

1. Most **"Normal"** animals are **owner surrendered**, indicating that pets are more likely to be healthy.

2. **"Aged"** animals often fall into the categories of **"Owner Surrender"** and **"Public Assistance,"** indicating that older animals are more likely to be surrendered or assisted.

3. **Wild animal admissions**: The heatmap shows that admissions of sick or injured wild animals occur more frequently in proportion to domestic animals.

4. **Euthanasia cases**: The condition **"Euthanasia Request"** indicates that animals in poor health are more likely to be surrendered with euthanasia requests.


### Let's check the relationship between Outcome Type and Intake Type .


```python
# Combine the intake and outcome datasets using 'Animal ID' as the key
combined_df = pd.merge(intakes_data, outcomes_data, on='Animal ID')

# Generate a contingency table to examine the relationship between Intake Condition and Outcome Type
relationship_table = pd.crosstab(combined_df['Intake Type'], combined_df['Outcome Type'])

percentage_table = relationship_table.div(relationship_table.sum(axis=1), axis=0)

plt.figure(figsize=(12, 12))
sns.heatmap(percentage_table, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Heatmap of Selected Features')

# Render the heatmap
plt.show()
```

### Does Age Play a Role in Adoption?

To determine if age affects adoption rates, we will divide the animals into two groups: **adopted** and **non-adopted**. By comparing the age distributions of these two groups, we can identify any significant differences in age.

#### Approach:

1. **Divide Data into Two Groups:**
   - Group 1: Animals that were adopted.
   - Group 2: Animals with all other outcomes (non-adopted).

2. **Statistical Metrics:**
   We will calculate the mean, median, and range of ages for both groups to see if there are noticeable differences.

3. **Visualizing Age Distributions:**
   Using box plots or violin plots, we will visualize the age distributions of adopted and non-adopted animals to look for trends.

This approach will help us see if younger or older animals are more likely to be adopted.


```python
age_of_adapted_animals =  np.array(outcomes_data[outcomes_data["Outcome Type"] == "Adoption"]["Age upon Outcome"].dropna().values)
age_of_nonadopted_animals =  np.array(outcomes_data[outcomes_data["Outcome Type"] != "Adoption"]["Age upon Outcome"].dropna().values)
```

```python
# Statistical metrics for both groups
mean_adopted = np.mean(age_of_adapted_animals)        
mean_nonadopted = np.mean(age_of_nonadopted_animals)   

median_adopted = np.median(age_of_adapted_animals)   
median_nonadopted = np.median(age_of_nonadopted_animals)  

min_adopted = np.min(age_of_adapted_animals)         
min_nonadopted = np.min(age_of_nonadopted_animals)   

max_adopted = np.max(age_of_adapted_animals)           
max_nonadopted = np.max(age_of_nonadopted_animals)  

# Print the results
print(f"Adopted Animals - Mean: {mean_adopted}, Median: {median_adopted}, Min: {min_adopted}, Max: {max_adopted}")
print(f"Non-Adopted Animals - Mean: {mean_nonadopted}, Median: {median_nonadopted}, Min: {min_nonadopted}, Max: {max_nonadopted}")
```

The difference in means and medians suggests that age might play a role in adoption outcomes. Adopted animals tend to be younger on average (mean: 573 days, median: 240 days) compared to non-adopted animals (mean: 868 days, median: 365 days). The max values also show that non-adopted animals can be older.

Next steps  involve visualizing the distributions.

```python
figure = plt.figure(figsize=(10, 6))
ax = figure.add_subplot()
ax.boxplot([age_of_adapted_animals, age_of_nonadopted_animals], 
           patch_artist=True, 
           labels=['Adopted', 'Non-Adopted'], 
           boxprops=dict(facecolor='lightblue', color='blue'),
           medianprops=dict(color='red', linewidth=2),showfliers=False) 
#  We don't show outliers because we're interested in the general trend.
ax.set_title('Age Distribution of Adopted vs Non-Adopted Animals', fontsize=16)
ax.set_ylabel('Age in Days', fontsize=14)
ax.set_yscale('log')

plt.show()
```

Conclusion: Yes, the age of the animal plays a significant role in determining the outcome type, particularly regarding adoption. The statistical analysis shows distinct differences in age distribution between adopted and non-adopted animals, suggesting that age is a relevant factor in the  adoption.

### Analyzing Animal Intake Trends Over Multiple Years

We will investigate whether animal intakes remain consistent throughout the year or fluctuate by comparing data across multiple years. Our approach:

- **Segment by Year**: We will analyze animal intake data for the  four random years.
- **Visualize Trends**: Plot the month  intake frequency for each year to identify patterns or anomalies.

This method will help us determine if there are consistent patterns or specific periods with more or fewer intakes.

```python
import random
random.seed(333)
years = random.sample(range(2013,2024),4)
years
```


This function visualizes the distribution of intakes across four randomly selected years. By analyzing intake frequency on a monthly basis, we can uncover trends and seasonal patterns that may not be evident when examining daily data.The monthly analysis allows us to smooth out daily fluctuations and focus on broader trends, making it easier to identify peak intake months and periods of lower activity. 



```python
def line_plot(years):
    figure = plt.figure(figsize=(12, 8))

    for i, year in zip(range(4), years):
        ax = figure.add_subplot(2, 2, i + 1) 
        intakes_frequency = list(intakes_data[intakes_data["DateTime"].dt.year == year]["DateTime"].dt.month.value_counts().sort_index().values)
        month = list(np.arange(1, len(intakes_frequency) + 1))
        ax.plot(month, intakes_frequency,color = '#D0021B')
        ax.set_title(f"Intakes Distribution in Year {year}")
        ax.set_xlabel("Month")
        ax.set_ylabel("Intakes Frequency")
        ax.grid(axis='both', linestyle="--")
        ax.set_xticks(np.arange(1,13))
    plt.tight_layout()
    plt.show()
```

```python
line_plot(years)
```

## Do intake conditions improve or worsen throughout the year (i.e., is it worse in winter compared to summer)?

```python
intakes_data["Intake Condition"].value_counts()
```

To answer the question, "Do intake conditions improve or worsen throughout the year?" we will divide the animals into two groups based on their intake condition:

1. **Normal Condition**: These animals are recorded as "Normal" upon intake.
2. **Non-Normal Condition**: This group will include animals categorized as injured, sick, or nursing. Other conditions, such as "Neonatal" or "Aged," are either too rare or may overlap with being sick or normal, so we'll exclude them from the main analysis for simplicity.

By visualizing these proportions using a line plot, we aim to determine whether more animals are admitted in worse conditions during certain parts of the year, particularly focusing on the comparison between winter and summer months.

```python
def line_plot_2(years):
    figure = plt.figure(figsize=(12, 8))

    for i, year in zip(range(4), years):
        ax = figure.add_subplot(2, 2, i + 1)
        
        normal_data = intakes_data[(intakes_data["DateTime"].dt.year == year) & (intakes_data["Intake Condition"] == "Normal")]
        normal_frequency = normal_data["DateTime"].dt.month.value_counts().sort_index().values
        sick_data = intakes_data[(intakes_data["DateTime"].dt.year == year) & 
                                 (intakes_data["Intake Condition"].isin(["Sick", "Injured", "Nursing"]))]
        sick_frequency = sick_data["DateTime"].dt.month.value_counts().sort_index().values
        months = np.arange(1, 13) 
        ax.plot(months[:len(normal_frequency)], normal_frequency, color='#1F77B4', label='Normal')
        ax.plot(months[:len(sick_frequency)], sick_frequency, color='#D0021B', label='Sick')
        ax.set_title(f"Intake Distribution in Year {year}")
        ax.set_xlabel("Month")
        ax.set_ylabel("Intake Frequency")
        ax.set_xticks(months)
        plt.grid(axis = 'x', linestyle='--', linewidth=0.5, color='gray')
        ax.legend(loc='upper left')

    plt.tight_layout()
    plt.show()
line_plot_2(years)
```

## Trends in Intake Conditions
* For Normal (Blue line):
In all years, the number of normal intakes generally increases towards summer and tends to peak around mid-year. After that, the number declines in the fall and winter (months 9 to 12).\
This suggests that normal conditions are better in summer (more intakes of healthy individuals) and worse in winter (fewer normal intakes).
* For Sick (Red line):
The trend for sick intakes is relatively stable throughout the year, with only slight fluctuations.
There seems to be a slight increase in sick cases in summer in some years, but it's not dramatic.
The trend does not indicate a major worsening of sick conditions in either summer or winter.
*General Interpretation:\
Normal intakes improve in summer (more healthy individuals), while they decline in winter.\
Sick intakes remain relatively constant throughout the year, with no significant worsening during winter compared to summer.
